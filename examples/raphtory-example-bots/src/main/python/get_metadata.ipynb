```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import ast
```

```python
data = pd.read_csv('/Users/pometry/Desktop/Features/sentiment.csv')
data.drop(columns=data.columns[0], axis=1, inplace=True)
data
```

```python
data.dropna(subset=["entities"], inplace=True)
data["entities"]=data["entities"].apply(lambda x:ast.literal_eval(x))
```

```python
def get_hashtags(h_list):
    if h_list == [] or h_list == None:
        return []
    else:
        return [x.get("text") for x in h_list]
```

```python
data["hashtags"]=data["entities"].apply(lambda x:get_hashtags(x.get("hashtags")))
data["hashtag_count"]=data["hashtags"].apply(lambda x:len(x))
```

```python
def get_user_mentions(h_list):
    if h_list == [] or h_list == None:
        return []
    else:
        return [x.get("id") for x in h_list]
    
data.iloc[1,3]
# get_user_mentions(data.iloc[1,3].get("user_mentions"))
```

```python
data["user_mentions"]=data["entities"].apply(lambda x:get_user_mentions(x.get("user_mentions")))
data["user_mentions_count"]=data["user_mentions"].apply(lambda x:len(x))
```

```python
data.drop("entities", axis=1, inplace=True)
```

```python
data
```

```python
data.dropna(subset=["public_metrics"], inplace=True)
data["public_metrics"]=data["public_metrics"].apply(lambda x:ast.literal_eval(x))
```

```python
data["retweets"]=data["public_metrics"].apply(lambda x:x.get("retweet_count"))
```

```python
data["likes"]=data["public_metrics"].apply(lambda x:x.get("like_count"))
```

```python
data.drop("public_metrics", axis=1, inplace=True)
data.drop("referenced_tweets", axis=1, inplace=True)
data.drop("source", axis=1, inplace=True)
```
