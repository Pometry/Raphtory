#### Setup

```python
import networkx as nx
from raphtoryclient import client
import pandas as pd
# Requires https://pygraphviz.github.io/documentation/stable/install.html
from networkx.drawing.nx_agraph import graphviz_layout
import matplotlib.pyplot as plt
```

#### Create Client

```python
raphtoryClient = client()
```

#### Connect to topic

```python
taintReader = raphtoryClient.createReader("TaintTracking", subscription_name='bindiff_reader')
```

#### Obtain and sort data from topic

```python
df_taint = raphtoryClient.getDataframe(taintReader)
# names=["runtime", "infected", "infectedBy", "infectionTime", "infectedAtHash", "value"])
#         result_0   result_1     result_2      result_3         result_4         result_5
df_taint = df_taint.sort_values("result_3")
df_taint
```

#### Create Graph from topic

```python
G = nx.DiGraph()
for i,row in df_taint.iterrows():
    G.add_edge(row['result_2'], row['result_1'])
G.number_of_edges(), G.number_of_nodes()
```

#### Draw graph with hierarchical layout

```python
nx.nx_agraph.write_dot(G,'taint_graph.dot')
pos=graphviz_layout(G, prog='dot')
nx.draw(G, pos, with_labels=False, arrows=True)
plt.savefig('taint_graph.png')
```
